This analysis will compare COVID-19 data in New Zealand to COVID-19 data in the United States in order to demonstrate statistical differences among nations containing the virus. 

## Data Source

The data used in this statistical analysis was collected from [European Centre for Disease Prevention and Control](https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide)

```python
%matplotlib inline
import pandas
```

## Importing Data from S3 Bucket

```python
df = pandas.read_excel('s3://ia241-kiersey/covid_data.xls')
df[:5]
```

## New Zealand and U.S Data

```python
New_Zealand_data= df.loc[df['countriesAndTerritories']=='New_Zealand']

New_Zealand_data[:10]
```

```python
US_data= df.loc[df['countriesAndTerritories']=='United_States_of_America']

US_data[:10]
```

Data above illustrates a significantly lower number of COVID-19 cases in New Zealand than the United States.

## Question 1. How do the cases in New Zealand compare to cases in the rest of the Oceania Continent?

```python
Oceania_data=df.loc[df['continentExp']=='Oceania']
Oceania_data[:10]
```

```python
Oceania_cases=Oceania_data.groupby('countriesAndTerritories').sum()['cases']
Oceania_cases.nlargest(15).plot.bar()
```

Q1. Answer: New Zealand placed fourth (behind Australia, French Polynesia, and Guam) in the highest number of positive COVID-19 cases

## Question 2. Which Month Recorded the Highest Number of  COVID-19 Cases in New Zealand and the U.S?

```python
sum_cases_by_month = df.groupby('month').sum()['cases']
sum_cases_by_month.nlargest(12).plot.bar()
```

Q2. Answer:The graph above demonstrates November (month 11) as reporting the most COVID-19 cases in both the United States and New Zealand

## Question 3. How Does the Total Number of Reported Cases Relate to the Total Number of Deaths in New Zealand in Comparison to that of the United States?

```python
New_Zealand_data.plot.scatter(x='cases',y='deaths',c='month')
```

The graph above illustrates the total number of New Zealand cases per month in 2020 and its relationship to the number of COVID reported deaths

```python
US_data.plot.scatter(x='cases',y='deaths',c='month')
```

Q3. Answer: The graph above illustrates the total number of U.S cases per month in 2020 and its relationship to the number of COVID reported deaths. As indicated in the data, the U.S had a significantly higher fatality rate among COVID-19 cases than New Zealand.
